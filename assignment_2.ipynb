```

def mean(score):
  row_mean=score.mean(axis=1,skipna=True)
  col_mean=score.mean(axis=0,skipna=True)
  overall_mean=col_mean.mean()
  print('\nRow wise mean')
  print(row_mean)
  print('\ncolumn wise mean')
  print(col_mean)
  print('\noverall  mean')
  print(overall_mean)

def median(score):
  row_median=score.median(axis=1,skipna=True)
  col_median=score.median(axis=0,skipna=True)
  overall_median=col_median.median()
  print('\nRow wise def median-->')
  print(  row_median)
  print('\ncolumn wise median-->')
  print(col_median)
  print('\noverall  median-->')
  print(overall_median)

def mode(score):
  row_mode=score.mode(axis=1)
  col_mode=score.mode(axis=0)
  overall_mode=col_mode.mode()
  print('\nRow wise mode')
  print(row_mode)
  print('\ncolumn wise mode')
  print(col_mode)
  print('\noverall  mode')
  print(overall_mode)

def sd(score):
 
  row_sd=score.std(axis=1)
  col_sd=score.std(axis=0)
  overall_sd=col_sd.std()
  print('\nRow wise standard deviation-->')
  print(row_sd)
  print('\ncolumn wise standard deviation-->')
  print(col_sd)
  print('\noverall  standard deviation-->')
  print(overall_sd)




mean(scores)
median(scores)
mode(scores)
sd(scores)
```
